# Data Preprocessing

## Importing Libs

```python
import numpy as np

import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

## Importing Dataset and Creating Dataset

```python

df = pd.read_csv('/Users/harmansingh/AI Engineering/Machine Learning/Basic Deep Learning/Artificial Neural Networks (ANN)/Churn_Modelling.csv')
df
```

```python
x= df.iloc[:,3:-1].values
y= df.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

## Encoding Categorical Data

### Label Encoding of Gender

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
x[:,2]= le.fit_transform(x[:,2])
print(x)
```

### One Hot Encoding on Geography

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
cl = ColumnTransformer( transformers = [('encoder',OneHotEncoder(),[1])], remainder = 'passthrough' )
x= np.array(cl.fit_transform(x))
print(x)
```

## Splitting Of Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Building ANN

## Initialize the ANN

```python
ann = tf.keras.models.Sequential()
```

## Adding Input Layer and First Hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation = 'relu'))
```

## Adding 2nd Hidden Layer

```python
ann.add( tf.keras.layers.Dense( units=6,activation = 'relu' ) )
```

## Adding Output Layer

```python
ann.add( tf.keras.layers.Dense( units=1,activation = 'sigmoid' ) )
```

# Training The ANN

## Compiling ANN

```python
ann.compile( optimizer = 'adam', loss= 'binary_crossentropy', metrics = ['accuracy'] )
```

## Training ANN

```python
ann.fit(x_train,y_train, batch_size=32, epochs =100)
```

# Making The Prediction and Evaluating The Model

## Predicting The Single Observation 

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
```

Therefore, our ANN model predicts that this customer stays in the bank!

Important note 1: Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting our values into a double pair of square brackets makes the input exactly a 2D array.

Important note 2: Notice also that the "France" country was not input as a string in the last column but as "1, 0, 0" in the first three columns. That's because of course the predict method expects the one-hot-encoded values of the state, and as we see in the first row of the matrix of features X, "France" was encoded as "1, 0, 0". And be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

## Predicting Test Set Result

```python
y_pred= ann.predict(x_test)
```

```python
y_pred= (y_pred>0.5)
print(np.concatenate(( y_test.reshape(len(y_test),1), y_pred.reshape(len(y_pred),1) ),1))
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
```

```python
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:',cm)
print()
print('Accuracy:', accuracy_score(y_test,y_pred) )
print()
print( 'Precision:',precision_score(y_test,y_pred) )
print()
print( 'Recall:',recall_score(y_test,y_pred) )
print()
print( 'F1_score:',f1_score(y_test,y_pred) )
```
